```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('worldbank_population_data.csv')
df.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = df.drop(df.columns[[2,3]], axis = 1)
df.head()
```

```python
df.isnull().sum()
```

```python
df.sort_values(by = '2023', ascending = False).head()[['Country Name', '2023']]
```

```python
df.sort_values(by = '2023', ascending = True).head()[['Country Name', '2023']]
```

```python
fig_2023 = px.choropleth(df,
                         locations = 'Country Name',
                         locationmode = 'country names',
                         color = '2023',
                         hover_name = 'Country Name',
                         title = '2023 Population Map',
                         color_continuous_scale = 'algae')

fig_2023.update_layout(height=700,
                  width=900)

fig_2023.show()
```

```python
df2 = pd.read_csv('continents2.csv')
df2.head()
```

```python
df2 = df2[['Country Name', 'Alpha-3','Continent']]
df2.rename(columns={"Alpha-3": "Country Code"}, inplace = True)
df2.head()
```

```python
df_v1 = df.merge(df2, on = 'Country Code', how = 'left')
```

```python
df_v1[df_v1.isna().any(axis=1)]
```

```python
df_v1.at[39, 'Continent'] = 'Europe'
df_v1.at[105, 'Continent'] = 'Europe'
```

```python
df_v1[df_v1.isna().any(axis=1)]
```

```python
df_v1.drop('Country Name_y', axis = 1, inplace = True)
df_v1.rename(columns={"Country Name_x": "Country"}, inplace = True)
```

```python
df_v1[df_v1.isna().any(axis=1)]
```

```python
df_v1.columns
```

```python
df_continents = df_v1.groupby('Continent')[['1960', '1961', '1962', '1963', '1964',
       '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973',
       '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982',
       '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991',
       '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000',
       '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
       '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
       '2019', '2020', '2021', '2022', '2023']].sum().sort_values(by = '2023', ascending = False)
```

```python
df_continents
```

```python
df_continents = df_continents.transpose()
df_continents.head()
```

```python
ax = df_continents.plot()
ax.set_title('Mean Population Growth')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment = 'right')
ax.set_ylabel('Population(billions)')
ax.legend(loc = 'center left', bbox_to_anchor = (1.0,0.5))
plt.show()
```

```python
df_v1['Population Change'] = df_v1['2023'] - df['1960']
```

```python
most_growth = df_v1.sort_values(by='Population Change', ascending=False).head(10)
most_decline = df_v1.sort_values(by='Population Change', ascending=True).head(10) 
```

```python
print("Countries with the most population growth from 1960 to 2023:")
most_growth[['Country', 'Population Change']]
```

```python
print("\nCountries with the most population decline from 1970 to 2022:")
most_decline[['Country', 'Population Change']]
```

```python
rural_pop = pd.read_csv('rural_population.csv')
rural_pop.head()
```

```python
urban_pop = pd.read_csv('urban_population.csv')
urban_pop
```

```python
#ser_df = pd.DataFrame(ser, columns = ['Technology'])
rural_pop_total = pd.DataFrame(rural_pop.iloc[:, 2:].sum().reset_index()).rename(columns={"index": "Year", 0:"Total Rural"})
rural_pop_total
```

```python
urban_pop_total = pd.DataFrame(urban_pop.iloc[:, 2:].sum().reset_index()).rename(columns={"index": "Year", 0:"Total Urban"})
urban_pop_total
```

```python
world_pop_total = pd.DataFrame(df.iloc[:, 2:].sum().reset_index()).rename(columns={"index": "Year", 0:"Total World"})
world_pop_total
```

```python
total_pop = rural_pop_total.merge(urban_pop_total, on = 'Year')
total_pop = total_pop.merge(world_pop_total, on = 'Year')
total_pop
```

```python
ax2 = total_pop.plot()
ax2.set_title('Total Population Growth')
plt.show()
```
